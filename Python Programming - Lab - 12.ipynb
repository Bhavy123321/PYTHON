### 08) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
x=range(1,10)
y=[2,1,4,5,7,8,9,6,3]
for i in range(len(y)):
    plt.text(i+1,y[i], y[i],horizontalalignment='center',verticalalignment='bottom')
plt.bar(x,y)
plt.show()
```

### 09) WAP create a Scatter Plot with several colors in Matplotlib?

```python
x=range(1,6)
y=[1,2,3,5,8]
colors=['red','yellow','green','blue','black']

plt.scatter(x,y,c=colors)
plt.colorbar()
plt.show()
```
